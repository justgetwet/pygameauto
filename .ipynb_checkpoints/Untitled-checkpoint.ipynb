```python
ua = {"User-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36 Edg/96.0.1054.57"}
from bs4 import BeautifulSoup
import pandas as pd
from pathlib import Path
import re
import requests

def get_dfs(url):
    res = requests.get(url, headers=ua)
    soup = BeautifulSoup(res.content, "html.parser")
    dfs = []
    if soup.find("table"):
        dfs = pd.io.html.read_html(soup.prettify())
    else:
        print(f"it's no table! {url}")
    return dfs

playerdetail = "https://www.oddspark.com/autorace/PlayerDetail.do?playerCd="

p_temp = Path("./photos/")
filenames = [p.name.strip(".jpg") for p in p_temp.iterdir() if p.is_file()]
racercode_d = {f.split("_")[1]: f.split("_")[0] for f in filenames}

def pred(entry_df):
    
    items = [item.replace("\u3000", " ") for item in entry_df.iloc[:,1].values]
    racers = ["".join(item.split()[:2]) for item in items]
    handis = [float(item.split()[0].strip("m"))/1000 for item in entry_df.iloc[:,3].values]

    time_data = []
    for racer, handi in zip(racers, handis):
        url = playerdetail + racercode_d[racer]
        dfs = get_dfs(url)
        df = [df for df in dfs if len(df.columns) == 12][0]
        times = []
        for row in df.itertuples():
            if type(row[7]) == str and row[7][0] == "良" and row[9] != "-":
                if type(row[9]) == str: 
                    time = float(re.sub("再| ","", row[9]))
                else:
                    time = row[9]
            times.append(time)  
        data = np.array(times) + handi
        time_data.append(data)
    
    wins, places = [], []
    for _ in range(8000):
        race_times = [np.random.choice(times) for times in time_data]
        m = min(race_times)
        w_indexes = [i+1 for i, x in enumerate(race_times) if x == m]
        w = np.random.choice(w_indexes)
        race_times[w-1] = 3.5 # dummy
        n = min(race_times)
        p_indexes = [i+1 for i, x in enumerate(race_times) if x == n]
        p = np.random.choice(p_indexes)
        wins.append(w)
        places.append(p)

    w_hist, _  = np.histogram(wins, bins=range(1, len(time_data)+2))
    w_rates = 100 * w_hist / w_hist.sum()
    p_hist, _  = np.histogram(places, bins=range(1, len(time_data)+2))
    p_rates = 100 * p_hist / p_hist.sum()

    res = []
    for i, (w, p) in enumerate(zip(w_rates, p_rates)):
        res.append((i+1, w, p))
        
    return res
```

```python
d={}
for i, row in enumerate(exac_df.values[2:,2:]):
    for j, value in enumerate(row):
        k = str(j+1) + ">" + str(i+1)
        d[k] = float(value)
d
```

```python
w = [itm[1] for itm in d.items() if itm[0][1] == "w"]
idx = range(1, len(w)+1)
sr_w = pd.Series(w, index=idx, name="win")
p1 = [itm[1] for itm in d.items() if itm[0][1:] == "p1"]
sr_p1 = pd.Series(p1, index=idx, name="p1")
p2 = [itm[1] for itm in d.items() if itm[0][1:] == "p2"]
sr_p2 = pd.Series(p2, index=idx, name="p2")
srs=[]
for i in idx:
    keys = [str(j) + ">" + str(i) for j in idx]
    vals = [d[key] for key in keys]
    sr = pd.Series(vals, index=idx, name="exa" + str(i))
    srs.append(sr)
df = pd.DataFrame(srs)
syn, syn2, syn3, syn4 = [], [], [], []
for i in range(len(idx)):
    arr = df.iloc[i,:].values
    nonan_list = arr[~np.isnan(arr)]
    a = [n for n in nonan_list if n != 0.]
    aa = np.array(a)
    a2 = np.array(sorted(a)[:2])
    a3 = np.array(sorted(a)[:3])
    a4 = np.array(sorted(a)[:4])
    o = 1/(1/aa).sum()
    o2 = 1/(1/a2).sum()
    o3 = 1/(1/a3).sum()
    o4 = 1/(1/a4).sum()
    syn.append(round(o, 1))
    syn2.append(round(o2, 1))
    syn3.append(round(o3, 1))
    syn4.append(round(o4, 1))
    
sr_syn = pd.Series(syn, index=idx, name="syna")
sr_syn2 = pd.Series(syn2, index=idx, name="syn2")
sr_syn3 = pd.Series(syn3, index=idx, name="syn3")
sr_syn4 = pd.Series(syn4, index=idx, name="syn4")
# print(keys)
data = [sr_w, sr_p1, sr_p2] + [sr_syn2, sr_syn3, sr_syn4, sr_syn] + srs
df = pd.concat(data, axis=1)
cmap = sns.color_palette("ch:s=-.25,rot=.25", as_cmap=True)
df.style.background_gradient(cmap=cmap).set_precision(1) #.set_properties(**{})
# df.style.applymap(lambda x: "background-color: gray" if x>10.0 else "background-color: none").set_precision(1)
# df
```
